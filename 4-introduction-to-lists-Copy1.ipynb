### Lists in Python

> A **list** is an ordered collection of data.  

Here is a list of restaurants.

````python
['chipotle', 'chopt', 'arbys']
````

Notice that we begin our list with the open bracket key `[`, which is located to the right of the letter `p` on the keyboard.  And we close our list with the closing bracket `]`.  After our open bracket we write our first element, `'chipotle'`.  
> An **element** is what we call each item in our list.

Each of our elements has to be a valid Python datatype, so we make each of our elements a string.  When we are done with our first element, `'chipotle'` we move onto a second element by writing a comma.  When we are done with each of our elements, we write the closing bracket `]`. 

Here is a list of numbers.

````python
[1, 2, 100]
````

### Selecting elements from a list

Let's go back to our list of restaurants and assign it to a variable called `restaurants`.

````python
restaurants = ['chipotle', 'chopt', 'arbys']
````

> Remember to press `shift + enter` to code along.

We can see the list just by typing the name of our variable.

````python
restaurants
````

Now we can ask different questions of this list.  For example, we can discover the number of elements in the list with the `len` function.

````python
len(restaurants)
````

And another thing we can do is select a given element from this list.  For example, we may want to select our first restaurant, chipotle.  Here's how we select chipotle from our list.

````python
restaurants[0]
````

And here's how we select chopt.

````python
restaurants[1]
````

> **Index**: These numbers, which indicate the location of each element in the list, are called the **index** of the element. 

So we select elements from a list based on their index, but the index always starts with the number 0.  So to select the first element from a list, we reference that list, then place the number 0 between opening and closing brackets.  

````python
restaurants = ['chipotle', 'chopt', 'arbys']
restaurants[0]
````

So the first element has an index of 0, and the second element has an index of 1.  

````python
restaurants[1]
````

Now because we may not want to always calculate the index of the last element, we can access the last element with `[-1]`. 

````python
restaurants[-1]
````

### Selecting Multiple Elements

So we just saw that we can select one element by using the bracket accessor.  Now let's try to select multiple elements.  Let's start with some restaurants.

````python
restaurants = ['chipotle', 'chopt', 'arbys', 'olive garden']
````

Now if we want to select `'chopt'` and `'arbys'` restaurants, we are essentially asking Python to start at the element with index 1, and stop *just before* the with index 3.

````python
restaurants[1:3]
````

This is called `slicing elements`.  So notice that we are **inclusive** of our starting element, and **exclusive** of our stopping element.  Also note that slicing does not affect our original list.

````python
restaurants
````

So if we want to hold onto just these middle restaurants, we must store our sliced elements in a new variable.

````python
mid_restaurants = restaurants[1:3]
mid_restaurants
````

We can also say to **start at an element and keep slicing until the end of a list** by just providing the starting index followed by a colon.

````python
restaurants[1:]
````

Or keep going **until an ending element** by just providing the stopping index.

````python
restaurants[:3]
````

We select elements from our list with the bracket accessors, and then placing the index that refers to our element.  The index just indicates the position of the element, starting with the number 0.
